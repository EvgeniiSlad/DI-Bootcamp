## 10

```python
plt.scatter(data['displacement'],data['acceleration'])
plt.xlabel('Displacement')
plt.ylabel('Acceleration')
```

## 11

```python
toyota = data[(data['car_name'].str.contains('toyota')) & (data['model_year'] > 78)]
toyota.head()
```

```python
not_toyota= data[(~data['car_name'].str.contains('toyota')) & (data['model_year'] > 78)]
not_toyota
```

```python
plt.bar(not_toyota['model_year'],not_toyota['cylinders'], label='Not toyota')
plt.bar(toyota['model_year'],toyota['cylinders'], label='Toyota')
plt.legend()
```
